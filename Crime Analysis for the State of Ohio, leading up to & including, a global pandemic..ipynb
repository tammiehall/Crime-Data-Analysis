```python
df_ohio.plot(kind='line', x='year', y='violent_crime', figsize=(8,6), color=['teal'])
plt.title('Total "Violent Crimes" in the State of Ohio 1979-2020')
plt.xlabel('Year')
plt.ylabel('Total Crimes');
```
